```python
import matplotlib.pyplot as plt

supplements = ["Whey protein", "Caffeine", "Casein", "Caffeine (2)", "BCAAs", "Beetroot juice"]
evidence = [6, 5, 5, 4, 4, 4]  
popularity = [8610, 7250, 2380, 7250, 5490, 267]  
colors = ["red", "blue", "green", "orange", "purple", "brown"]

plt.figure(figsize=(8, 6))
plt.scatter(evidence, popularity, color=colors, edgecolors='black', s=100)

for supp, x, y, c in zip(supplements, evidence, popularity, colors):
    plt.text(x, y, supp, fontsize=10, ha='right', color=c)

plt.xlim(0, max(evidence) + 1)  
plt.ylim(0, max(popularity) + 400)  

plt.xlabel("Evidence Level Score")
plt.ylabel("Popularity")
plt.title("Supplement Popularity vs. Evidence Level")
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```

```python
import matplotlib.patheffects as path_effects  

supplements = ["Whey protein", "Caffeine", "Casein", "Caffeine (2)", "BCAAs", "Beetroot juice"]
evidence = [6, 5, 5, 4, 4, 4]  
popularity = [8610, 7250, 2380, 7250, 5490, 267]  
colors = ["red", "blue", "green", "orange", "purple", "brown"]

plt.figure(figsize=(8, 6))
plt.scatter(evidence, popularity, color=colors, edgecolors='black', s=100)

for supp, x, y in zip(supplements, evidence, popularity):
    text = plt.text(x, y, supp, fontsize=9, ha='left', color='black', fontweight='bold')

    text.set_path_effects([path_effects.Stroke(linewidth=1, foreground='white'),
                           path_effects.Normal()])

plt.xlim(0, max(evidence) + 1.1)  
plt.ylim(0, max(popularity) + 400)  

plt.xlabel("Evidence Level Score")
plt.ylabel("Popularity")
plt.title("Supplement Popularity vs. Evidence Level")
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```
